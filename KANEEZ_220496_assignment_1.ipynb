# Test function to validate your model architecture

```python
import torch

def validate_model(pythia_model,text):
  # Tokenize input
  tokens = pythia_model.to_tokens(text)

  # Dictionary to store hook outputs
  hook_outputs = {}

  # Define hooks to capture intermediate activations
  hook_map = {
      "blocks.0.hook_resid_pre": "hook_resid_pre",
      "blocks.0.ln1.hook_normalized": "ln1.hook_normalized",
      "blocks.0.hook_attn_out": "hook_attn_out",
      "blocks.0.hook_resid_post": "hook_resid_post",
      "blocks.0.hook_mlp_out": "hook_mlp_out",
  }

  # Register hooks
  fwd_hooks = [(hook_name, lambda tensor, hook, name=short_name: hook_outputs.update({name: tensor.detach()}))
              for hook_name, short_name in hook_map.items()]

  # Run model with hooks
  pythia_model.run_with_hooks(tokens, fwd_hooks=fwd_hooks, return_type="logits")

  # Extract embeddings
  embedding = pythia_model.embed(tokens)
  residual = embedding

  # Verify stored activations
  for key in hook_map.values():
      stored_tensor = hook_outputs[key]
      computed_tensor = None

      if key == "hook_resid_pre":
          computed_tensor = residual
      elif key == "ln1.hook_normalized":
          computed_tensor = pythia_model.blocks[0].ln1(residual)
      # elif key == "hook_attn_out":
      #     computed_tensor = pythia_model.blocks[0].attn(residual)
      elif key == "hook_mlp_out":
          computed_tensor = pythia_model.blocks[0].mlp(residual)
      # elif key == "hook_resid_post":
      #     computed_tensor = residual + computed_tensor

      # Print verification results
      print(f"\nVerifying '{key}':")
      # print(f"Computed shape: {computed_tensor.shape}")
      print(f"Stored shape: {stored_tensor.shape}")
      # match = torch.equal(computed_tensor, stored_tensor)
      # print(f"Exact match: {match}")

  # Final verification check
  all_correct = all(torch.equal(hook_outputs[key], computed_tensor) for key in hook_map.values())
  print("\nFinal verification result:", all_correct)
```

```python
text = "Why did the chicken cross the road"
all_verified = validate_model(pythia_model, text)
if all_verified:
  print("The model architecture is correct.")
else:
  print("The model architecture is incorrect.")
```
